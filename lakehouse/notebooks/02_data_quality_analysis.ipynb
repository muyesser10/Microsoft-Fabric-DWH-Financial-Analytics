# Data Quality Analysis Notebook

## Purpose

This notebook evaluates the quality of the dataset before performing transformations.

The analysis focuses on identifying common data issues such as:

- Missing values
- Duplicate records
- Data type inconsistencies
- High-cardinality categorical columns

Understanding these issues allows us to design an effective data cleaning strategy.

## Dataset

Input dataset:

Files/exim_authorizations.csv

## Output

A set of data quality insights used to guide the data cleaning process.

```python
from pyspark.sql.functions import col, count, when
```

```python
df = spark.read.format("csv") \
    .option("header","true") \
    .load("Files/raw/Exim Financial Dataset.csv")
```

```python
missing_values = df.select([
    count(when(col(c).isNull(), c)).alias(c)
    for c in df.columns
])

display(missing_values)
```

```python
df.dropDuplicates().count()
```
